Préparation et chargement des données, describe du X_train

```python
X_train_path = '../data/X_train_78VdSWL.csv'
Y_train_path = '../data/y_train_u0UkKEh.csv'
X_test_path = '../data/X_test_XKVc4no.csv'
Y_test_path =  '../data/new_output_sample.csv'
X_test_path = '../data/X_test_XKVc4no.csv'
X_train = pd.read_csv(X_train_path)
X_test = pd.read_csv(X_test_path)
y_train = pd.read_csv(Y_train_path)
X_train.describe(include='all')


```

On va extraire les colonnes holed_, c'est à dire celles avec des trous

```python
no_holes = [col for col in X_train.columns if not col.startswith('holed_')]
holed = [col for col in X_train.columns if col.startswith('holed_')]
```

On va entraîner une régression linéaire PAR colonne avec des trous, et les ranger dans un dico pour pouvoir y accéder rapidement, afin d'éviter de supprimer toute les colonnes avec des NaN alors qu'elles contiennent de l'infromation sur les autres colonnes  qui pourornt être utilisées pour les autres régressions, on va remplir les NaN par la valuer moyenne de la colonne.


```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

X_train_filled = X_train[holed].fillna(X_train[holed].mean())
y_train_filled = y_train[holed].fillna(y_train[holed].mean())
X_test_filled = X_test[holed].fillna(X_test[holed].mean())


#scaler = StandardScaler()

#X_train_scaled = scaler.fit_transform(X_train_filled[holed])
#X_train_scaled = pd.DataFrame(X_train_scaled, columns=holed, index=X_train_filled.index)

#X_test_scaled = scaler.transform(X_test_filled[holed])
#X_test_scaled = pd.DataFrame(X_test_scaled, columns=holed, index=X_test_filled.index)

modeles = {}
for col in holed:
    X = X_train_filled[holed]
    y = y_train_filled[col]

    model = LinearRegression()
    model.fit(X, y)
    modeles[col] = model
```

On applique le modele sur X_test


```python

y_pred = pd.DataFrame(index=X_test.index)

predictions = []
for col in holed:
    preds = modeles[col].predict(X_test_filled)
    predictions.append(pd.Series(preds, index=X_test.index, name=col))

y_pred = pd.concat(predictions, axis=1)
```

```python
y_pred.to_csv('new_output_sample.csv', index=False)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score


Y_test_path =  '../data/new_output_sample.csv'
y_test = pd.read_csv(Y_test_path)

for col in holed:
    mae = mean_absolute_error(y_test[col], y_pred[col])
    score = r2_score(y_test[col], y_pred[col])
    print(f"MAE et R2 pour {col} : {mae}; {score}")
```

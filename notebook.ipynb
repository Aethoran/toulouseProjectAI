On va extraire les colonnes holed_, c'est à dire celles avec des trous

```python
no_holes = [col for col in X_train.columns if not col.startswith('holed_')]
holed = [col for col in X_train.columns if col.startswith('holed_')]
```

On va entraîner une régression linéaire PAR colonne avec des trous, et les ranger dans un dico pour pouvoir y accéder rapidement, afin d'éviter de supprimer toute les colonnes avec des NaN alors qu'elles contiennent de l'infromation sur les autres colonnes  qui pourornt être utilisées pour les autres régressions, on va remplir les NaN par la valuer moyenne de la colonne.


```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

X_train_nodate = X_train.copy()
X_test_nodate = X_test.copy()
y_train_nodate = y_train.copy()

X_train_nodate['Horodate'] = pd.to_datetime(X_train_nodate['Horodate'])
X_test_nodate['Horodate'] = pd.to_datetime(X_test_nodate['Horodate'])
y_train_nodate['Horodate'] = pd.to_datetime(y_train_nodate['Horodate'])

for df in [X_train_nodate, X_test_nodate, y_train_nodate]:
    df['year'] = df['Horodate'].dt.year
    df['month'] = df['Horodate'].dt.month
    df['day'] = df['Horodate'].dt.day
    df['hour'] = df['Horodate'].dt.hour
    df['weekday'] = df['Horodate'].dt.weekday

X_train_nodate = X_train_nodate.drop(columns=['Horodate'])
X_test_nodate = X_test_nodate.drop(columns=['Horodate'])
y_train_nodate = y_train_nodate.drop(columns=['Horodate'])


X_train_filled = X_train_nodate.copy()
X_test_filled = X_test_nodate.copy()

X_train_filled = X_train_nodate.fillna(X_train_nodate.mean())
X_test_filled  = X_test_nodate.fillna(X_test_nodate.mean())

```

```python
""" scaler = StandardScaler()
X_train_scaled = pd.DataFrame(
    scaler.fit_transform(X_train_filled), 
    columns=X_train_filled.columns, 
    index=X_train_filled.index)

X_test_scaled = pd.DataFrame(
    scaler.transform(X_test_filled),
    columns=X_test_filled.columns,
    index=X_test_filled.index) """

X_test_filled = X_test_nodate.reindex(columns=X_train_nodate.columns)
print(X_train_filled.columns.equals(X_test_filled.columns))
X_test_filled = X_test_filled.fillna(0)
```

```python
models = {}
pred_list = []

for col in holed:
    mask = ~y_train_nodate[col].isna()
    X = X_train_filled.loc[mask, :]
    y = y_train_nodate.loc[mask, col]

    model = LinearRegression()
    model.fit(X, y)
    models[col] = model

    pred = pd.DataFrame(model.predict(X_test_filled), columns=[col], index=X_test.index)
    pred_list.append(pred)

y_pred = pd.concat(pred_list, axis=1)
```

```python
y_pred_file = pd.concat(pred_list, axis=1)
y_pred_file.to_csv('new_output_sample.csv', index=False)
```

On applique le modele sur X_test


```python
predictions = []
for col in holed:
    preds = models[col].predict(X_test_filled)
    predictions.append(pd.Series(preds, index=X_test.index, name=col))

y_pred = pd.concat(predictions, axis=1)
```

```python
y_pred.to_csv('new_output_sample.csv', index=False)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score


Y_test_path =  '../data/new_output_sample.csv'
y_test = pd.read_csv(Y_test_path)


for col in holed:
    if col in y_test.columns:
        mae = mean_absolute_error(y_test[col], y_pred[col])
        r2 = r2_score(y_test[col], y_pred[col])
        print(f"MAE et R² pour {col} : {mae:.4f} ; {r2:.4f}")

```
